```python
import pandas as pd
df = pd.read_csv('training.csv', header=None)
df
```

```python
df = df.rename(columns={0 : "id", 1 : "premise"})
df
```

```python
df = df.drop(df.columns[0], axis=1)
df
```

```python
df_1 = pd.read_csv('actual_training.csv', sep=" ", header=None)
df_1
```

```python
df_1 = df_1.rename(columns={0:"type", 1:"noun1", 2:"noun2"})
df_1["canonical_form"] = (df_1["type"] + " " + df_1["noun1"] + " " + df_1["noun2"])
df_1
```

```python
df_1 = df_1.drop(df_1.columns[[0, 1, 2]], axis = 1)
df_1
```

```python
df_train = pd.DataFrame(columns=["premise", "canonical_form"])

for i, row1 in df.iterrows():
    premise = row1["premise"]  
    for m, row2 in df_1.iterrows():
        canonical = row2["canonical_form"]
        c = canonical.split(" ")  
        if len(c) >= 3 and c[1] in premise and c[2] in premise:
            new_row = pd.DataFrame({"premise": [premise],"canonical_form": [canonical]})
            df_train = pd.concat([df_train, new_row], ignore_index=True)
            break  

df_train
```

```python
df_train.rename(columns={"premise" : "input", "canonical_form" : "output"}, inplace=True)
df_train.to_json("training_t5.jsonl", orient="records", lines=True)
```
