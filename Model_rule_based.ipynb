```python
import spacy
import nltk
sp = spacy.load('en_core_web_sm')
def remove_punct(text):
    doc = sp(text)
    tokens = []
    for token in doc:
        if token.is_punct:
            if token.text == "." or token.text == ";" or token.text == ":":
                tokens.append(token.text)
        else:
            tokens.append(token.text)
    return " ".join(tokens)
df["syllogism"] = df["syllogism"].astype('str').apply(remove_punct)
df
```

```python
import spacy

# -----------------------------
# Load NLP
# -----------------------------
nlp = spacy.load("en_core_web_sm")

# -----------------------------
# 19 Valid Syllogisms
# -----------------------------
VALID_SYLLOGISMS = {
    1: {"AAA", "EAE", "AII", "EIO"},
    2: {"EAE", "AEE", "EIO", "AOO"},
    3: {"AAI", "IAI", "AII", "EAO", "OAO", "EIO"},
    4: {"AAI", "AEE", "IAI", "EAO", "EIO"},
}

# -----------------------------
# Utilities
# -----------------------------
def resolve_predicate(token):
    """Resolve 'type/kind of X' → X."""
    for c in token.children:
        if c.dep_ == "prep" and c.lemma_ == "of":
            for g in c.children:
                if g.dep_ == "pobj":
                    return g
    return token

def extract_nouns(doc):
    return [t.lemma_ for t in doc if t.pos_ == "NOUN"]

def is_existential(doc):
    return any(t.lemma_ in {"exist"} for t in doc) or \
           any(t.lemma_ == "there" for t in doc)

# -----------------------------
# Logic Extractor (NEVER None)
# -----------------------------
def extract_logic(sentence):
    doc = nlp(sentence)

    # --- Existential cases ---
    if is_existential(doc):
        nouns = extract_nouns(doc)
        if len(nouns) >= 2:
            return ("I", nouns[0], nouns[1])
        elif len(nouns) == 1:
            return ("I", nouns[0], nouns[0])
        else:
            return ("I", "thing", "thing")

    # --- Copular cases ---
    copula = None
    for t in doc:
        if t.dep_ == "ROOT" and t.lemma_ == "be":
            copula = t
            break

    if copula:
        subject = None
        predicate = None
        quantifier = None
        negated = False

        # subject
        for c in copula.children:
            if c.dep_ == "nsubj":
                subject = c
                break

        # predicate
        for c in copula.children:
            if c.dep_ in {"attr", "acomp"}:
                predicate = resolve_predicate(c)
                break

        # quantifier
        if subject:
            for c in subject.children:
                if c.dep_ == "det":
                    if c.lemma_ in {"all", "every", "each"}:
                        quantifier = "A"
                    elif c.lemma_ == "no":
                        quantifier = "E"
                    elif c.lemma_ in {"some", "a", "an"}:
                        quantifier = "I"

        # negation
        for c in copula.children:
            if c.dep_ == "neg":
                negated = True

        if quantifier == "A" and negated:
            quantifier = "O"
        if quantifier == "I" and negated:
            quantifier = "O"

        if quantifier and subject and predicate:
            return (quantifier, subject.lemma_, predicate.lemma_)

    # --- Fallback (guaranteed) ---
    nouns = extract_nouns(doc)
    if len(nouns) >= 2:
        return ("I", nouns[0], nouns[1])
    elif len(nouns) == 1:
        return ("I", nouns[0], nouns[0])
    else:
        return ("I", "thing", "thing")

# -----------------------------
# S, P, M identification
# -----------------------------
def identify_terms(p1, p2, concl):
    _, s1, p1_ = p1
    _, s2, p2_ = p2
    _, s3, p3_ = concl

    S = s3
    P = p3_
    premise_terms = {s1, p1_, s2, p2_}
    candidates = premise_terms - {S, P}
    if len(candidates) != 1:
        return None
    M = candidates.pop()
    return {"S": S, "P": P, "M": M}

# -----------------------------
# Figure detection
# -----------------------------
def determine_figure(p1, p2, terms):
    _, s1, p1_ = p1
    _, s2, p2_ = p2
    S, M = terms["S"], terms["M"]

    def form(s, p):
        return ("M" if s == M else "P", "M" if p == M else "P")

    p1f = form(s1, p1_)
    p2f = ("S" if s2 == S else "M", "M" if p2_ == M else "P")

    if p1f == ("M", "P") and p2f == ("S", "M"): return 1
    if p1f == ("P", "M") and p2f == ("S", "M"): return 2
    if p1f == ("M", "P") and p2f == ("M", "S"): return 3
    if p1f == ("P", "M") and p2f == ("M", "S"): return 4
    return None

def extract_mood(p1, p2, concl):
    return p1[0] + p2[0] + concl[0]

# -----------------------------
# Validity Predictor (THE MODEL)
# -----------------------------
def predict_validity(p1, p2, concl):
    terms = identify_terms(p1, p2, concl)
    if terms is None:
        return False
    figure = determine_figure(p1, p2, terms)
    if figure is None:
        return False
    mood = extract_mood(p1, p2, concl)
    return mood in VALID_SYLLOGISMS.get(figure, set())

# -----------------------------
# Apply to DataFrame
# -----------------------------
# Assumes columns: premise1, premise2, conclusion
# Produces: logical_expression_1/2/3 and predicted_validity

def run_pipeline(df):
    df = df.copy()
    df["logical_expression_1"] = df["premise1"].apply(extract_logic)
    df["logical_expression_2"] = df["premise2"].apply(extract_logic)
    df["logical_expression_3"] = df["conclusion"].apply(extract_logic)

    df["predicted_validity"] = df.apply(
        lambda r: predict_validity(
            r["logical_expression_1"],
            r["logical_expression_2"],
            r["logical_expression_3"]
        ),
        axis=1
    )
    return df
```
